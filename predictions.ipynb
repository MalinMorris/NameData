# Predicting Rankings 2024
I will attempt to use the trends of the top 12 boy and girl names in 2023 to predict the 2024 rankings for the top 10 boy and girl names.

```python
import pandas as pd
import matplotlib.pyplot as plt
import functions as f
min_data_year = 1880
max_data_year = 2025
female_color = 'xkcd:pink'
male_color = 'xkcd:light blue'
default_color = 'xkcd:periwinkle'
default_color2 = 'xkcd:grass green'
colors = ['xkcd:red','xkcd:orange','xkcd:mustard','xkcd:grass green', 'xkcd:light blue','xkcd:cerulean','xkcd:indigo', 
          'xkcd:lavender', 'xkcd:pink', 'xkcd:black', 'xkcd:tan', 'xkcd:grey blue']
```

```python
data = pd.read_csv("SSANameData.txt", index_col = 0)
data = data.drop('index', axis = 1)
girls_2023 = ['Olivia', 'Emma', 'Charlotte', 'Amelia', 'Sophia', 'Mia', 'Isabella', 'Ava', 'Evelyn', 'Luna',
            'Harper', 'Sofia']
boys_2023 = ['Liam', 'Noah', 'Oliver', 'James', 'Elijah', 'Mateo', 'Theodore', 'Henry', 'Lucas', 'William',
             'Benjamin', 'Levi']
data.head()
```

```python
color_index = 0
for n in girls_2023:
    name_counts, ignore = f.name_counts_years(n.lower(), 'f', 2000, 2024)
    plt.plot(range(2000,2024), name_counts, 
             color = colors[color_index], label = f"{n}, (f)")
    color_index += 1
plt.title("Top 12 Girl Names 2023")
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid()
plt.legend()
```

```python
color_index = 0
for n in boys_2023:
    name_counts, ignore = f.name_counts_years(n.lower(), 'm', 2000, 2024)
    plt.plot(range(2000,2024), name_counts, 
             color = colors[color_index], label = f"{n}, (m)")
    color_index += 1
plt.title("Top 12 Boy Names 2023")
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid()
plt.legend()
```

## Predictions
Throughout the previous generations, the popular girls names are far more indicative of a certain time, so girls should be easy. The top 2 will definitely remain Olivia and Emma for a little while, but I think that Amelia has the right sounds to become a Gen Beta defining name and will jump to 3, pushing Charlotte to 4. Sophia and Mia are fairly close and could switch, given Mia is trending upward, but I think they will remain the same. Isabella and Ava are both trending downward while Evelyn is fairly steady, but Ava is moving faster. I think 7, 8, 9 will be Isabella, Evelyn, Ava. For the tenth name, Harper has been downtrending, and as I've read in name forums, Luna seems to have had its run as a blip in the top 10. Since Sofia has been increasing, I think it can jump into the top 10.<br>

Boys names are going to be more difficult to predict. The top three will certainly remain Liam, Noah, Oliver, but the others are so close together. Mateo has really had a sharp rise the last few years, and I think it could jump to 4. James and Elijah are very close but I think they will stay that order in 5 and 6. While Theodore has been increasing, it flattened out in 2023, so I think it will stay at 7. Henry and Lucas only differ by 106 in 2023, but given Henry just started to decrease, I think Lucas will move up one. Finally, there is more of a difference between William, Benjamin, and Levi, so I believe William will stay 10.
